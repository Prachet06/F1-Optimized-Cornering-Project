## Initial Data Exploration
In the python cell below I will load the JSON file with data from the session I am interested in and check what data type it is laoded as:

```python
import json
with open("../Data/sector-time-data/session-14.json") as f:
    session_info = json.load(f)

print(type(session_info))
```

Since the loaded JSON file is a dictionary I will check what its keys are below:

```python
print(session_info.keys())
```

The key whose value I am interested in is: "classification-data", below I will extract the values associated with this key:

```python
session_data = session_info["classification-data"]
print(type(session_data))
print("Length: ", end = "")
print(len(session_data))
```

Since the data type of the session_data is a list with just one element, I will attempt to extract useful information from it in the next code chunk:

```python
session_data = session_info["classification-data"][0]
print(type(session_data))
```

The first and only element is a dictionary. Therefore, I will check what its keys are:

```python
print(session_data.keys())
```

The key "lap-time-history" is of interest. I will check what its data type is:

```python
laps = session_data["lap-time-history"]
print(type(laps))
```

This is another dictionary, I will check its keys below:

```python
print(laps.keys())
```

The key of interest here is: "lap-history-data", I will check what the values associated with it are:

```python
laps_hist = laps["lap-history-data"]
print(type(laps_hist))
```

It is a list, I will look into what is stored at each of its indexes below:

```python
print(laps_hist[0])
print(type(laps_hist[0]))
```

It looks like each index contains each lap's data in the form of a dictionary. For the first lap, I will check which key contains sector times below:

```python
lap_1 = laps_hist[0]
print(lap_1.keys())
```

The key I am interested in is: "sector-1-time-in-ms". I now know what is needed to extract each lap's sector times, I will now write a loop to extract the sector times for each recorded lap and print it:

```python
for i in range (len(laps_hist)):
    current_lap = laps_hist[i]
    print(f"Lap {i+1} Sector 1 Time in ms: {current_lap["sector-1-time-in-ms"]}")
```
